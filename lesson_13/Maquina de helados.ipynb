<h4>Máquina automática de vender helados</h4>
<p>
Te han contratado en una empresa que se dedica a la fabricación de máquinas de helados para la venta automatizada en oxxos, centros comerciales y plazas. Te han asignado como el programador de la máquina de helados. 
</p>

</center>

<h4>Requisitos del programa de máquinas de hacer helados</h4>
<hr>
<p>
1. Indique al usuario preguntando “¿Qué helado le gustaría? (fresa/ chocolate/ combinado): "
Preguntar al usuario si ¿Le gustaría agregar jarabe a su helado?, también preguntar al usuario si ¿Le gustaría agregar chispas a su helado? 

a. Verifique la entrada del usuario para decidir qué hacer a continuación. 
b. El mensaje debe mostrar cada vez que se ha completado la acción, p. Ej. una vez que un helado es
dispensado. El mensaje debería aparecer de nuevo para atender al próximo cliente.
</p>
<hr>
<p>
2. Apague la máquina de helado ingresando "apagado" en el mensaje.
Los encargados del mantenimiento de la máquina de helado pueden utilizar "apagado" como palabra secreta para apagar
la máquina. Su código debería finalizar la ejecución cuando esto suceda.
<p>
<hr>
<p>
3. Imprimir informe.
a. Cuando el usuario ingresa "informe" al mensaje, se debe generar un informe que muestre
los valores actuales de los recursos. Por ejemplo:
<ul>
    <li>fresa: 300 ml</li>
    <li>chocolate: 450 ml</li>
    <li>jarabe: 200 ml</li>
    <li>chispas: 250 grs</li> 
    <li>conos: 65</li>
    <li>dinero: &#36; 162.</li>
</ul>
</p>
<hr>
4. ¿Verificar recursos suficientes?
Cuando el usuario elige un helado, el programa debe verificar si hay suficientes recursos para hacer ese helado.
Por ejemplo, si requiere la opción 1 debe dar 100 ml de helado fresa  pero solo quedan 50 ml en la máquina, no debería
continuar haciendo el helado, debe informar al cliente: "Lo siento, no hay suficiente helado de fresa". Lo mismo debería suceder si se agota otro recurso, por ejemplo los conos y el helado de chocolate, si no hay chispas o el jarabe se debe indicar que no serán agregados.  
</p>
<hr>
<p>
Si hay suficientes recursos para hacer el helado seleccionado, entonces el programa debe pedir al usuario que inserte monedas.
Las monedas que puede insertar el usuario son de 1,2,5 y 10 pesos mexicanos.
<center>
<img src="https://thumbs.dreamstime.com/z/monedas-mexicanas-56533858.jpg" width="350">
</center>
Calcula el valor monetario de las monedas insertadas. Por ejemplo. 1 peso, 2 pesos y 5 pesos total = &#36; 8 pesos.
</p>
<hr>
<p>
6. ¿Comprobar que la transacción se realizó correctamente?
Compruebe que el usuario haya introducido suficiente dinero para comprar el helado que seleccionó. Por ejemplo, un helado de fresa cuesta &#36;10 pesos, pero si agrega chispas cuesta 11 pesos, pero solo insertaron 8 pesos y luego de contar las monedas, el programa debería decir: Lo siento, no es suficiente dinero. Dinero reembolsado.Pero si el usuario ha introducido suficiente dinero, entonces el costo del helado se agrega al
máquina como ganancia y esto se refleja la próxima vez que se active el "informe". Por ejemplo. 
<ul>
    <li>fresa: 200 ml</li>
    <li>chocolate: 450 ml</li>
    <li>jarabe: 200 ml</li>
    <li>chispas: 230 grs</li> 
    <li>conos: 64</li>
    <li>dinero: &#36; 173.</li>
</ul>
Si el usuario ha introducido demasiado dinero, la máquina debe ofrecer cambio.
Por ejemplo "Aquí hay &#36;3 pesos en cambio". 
</p>
<hr>
<p>
7. Prepara helado.
Si la transacción es exitosa y hay suficientes recursos para hacer el helado he ha seleccionado por el usuario, los ingredientes para hacer la  deben deducirse de la máquina de helados los recursos. Se deben reflejar en el reporte los recursos gastados y el dinero ganado. 
Recursos antes:
<ul>
    <li>fresa: 300 ml</li>
    <li>chocolate: 450 ml</li>
    <li>jarabe: 200 ml</li>
    <li>chispas: 250 grs</li> 
    <li>conos: 65</li>
    <li>dinero: &#36; 162.</li>
</ul>
Recursos ahora:
<ul>
    <li>fresa: 200 ml</li>
    <li>chocolate: 450 ml</li>
    <li>jarabe: 200 ml</li>
    <li>chispas: 230 grs</li> 
    <li>conos: 64</li>
    <li>dinero: &#36; 173.</li>
</ul>

Una vez que se hayan deducido todos los recursos, dígale al usuario “Aquí está su helado de fresa  ¡Disfrutar!". Si
el helado de fresa fue el seleccionado.
</p>
