<h4>El juego de piedra, papel o tigera</h4>
<img src = "https://upload.wikimedia.org/wikipedia/commons/1/1d/Piedra_papel_tijeras.jpg" width = "400">
<br>
<p>
Realizar un programa que simule un juego de piedra, papel o tijera con la computadora. 
</p>

```python
#importar el modulo random
import random 

piedra = """
Piedra
    _______
___'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
"""
papel = """
Papel
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
"""

tigeras = """
Tigeras 
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
"""
#lista con jugadas posibles
#            0      1      2
jugadas = [piedra,papel,tigeras]
#selección de jugada por el usuario 
jugada_usuario = int(input('¿Qué jugada eliges?,  Escribe 1 para piedra, 2 para papel o 3 para tijeras\n'))
#ejecución de la partida 
if jugada_usuario>3 or jugada_usuario<1:
    print("Jugada no valida, reinicia el juego")
else:
    jugada_usuario -= 1
    print("Tu jugada es ▶")
    print(jugadas[jugada_usuario])
    #computar la jugada del ordenador
    jugada_maquina = random.randint(0,2)
    print("La jugada de la computadora es ▶")
    print(jugadas[jugada_maquina])
    #seleccionar el ganar
    if jugada_usuario == 0 and  jugada_maquina == 2:
        print("Ganaste 😎")
    elif jugada_usuario == 2 and  jugada_maquina == 0:
        print("Perdiste 😭")
    elif jugada_usuario > jugada_maquina:
        print("Ganaste 😎")
    elif jugada_usuario < jugada_maquina:
        print("Perdiste 😭")
    elif jugada_usuario == jugada_maquina:
        print("Empate 🤝")
```
