<h4>Estructura de control repetitiva <b>mientras</b> </h4>
<p>Esta estructura repite una acción o grupo de acciones mientras una expresión lógica sea cierta.</p>
    <img src="./img/while.png" width="300">

```python
#ejemplo de while imprime los primeros n numeros 1,2,3,...,n
n = int(input('¿Cuántos numeros desea mostrar? \n'))
i = 1
while i<=n:
    print(i, end=",")
    i += 1
```

```python
#ejemplo de while suma de n numeros 1+2+3+...+n'
n = int(input('Ingrese el limite de la suma n\n'))
suma = 0
i = 1
while i<=n:
    suma += i
    i += 1
print(f'La suma es {suma}')
```
