<h4>Aprobaste o reprobaste</h4>
<p>
Tienes  las calificaciones  de estudiantes de tu materia de matem치ticas, debes desarrollar un programa que pueda determinar quien no aprob칩 la materia con menos de 70, quien tiene m치s de 70 para aprobar la materia. 
Cuando tu alumno te indique su nombre debes decirle si aprobo o reprobo. 
</p>

```python
#calificaciones de tus alumnos de matem치ticas 
calificaciones_mate = {
'HERNANDEZ LUIS' : 50, 
'TORRES LAURA': 100,
'GONZALEZ ALVARO': 100,
'SALINAS CARLOS': 80,
'AGUILAR JOSE': 69, 
'CRUZ MARIA': 55, 
'GARCIA JESUS':92,
'AGUIRRE GERARDO': 95,
'GOMEZ JOSEFINA': 82, 
'VARELA JORGE': 99,
'ROSAS ERICK': 89,
}
```

```python
#agrupar estudiantes en aprobados y reprobados 
estudiantes = {'aprobados': [], 'reprobados':[]}
for estudiante in calificaciones_mate:
    if calificaciones_mate[estudiante] > 70:
        estudiantes['aprobados'].append(estudiante)
    else:
         estudiantes['reprobados'].append(estudiante)
print(estudiantes)
```

```python
#Informar al estudiante si aprobo o reprobo 
nombre = input('Alumno ingrese su apellido paterno y primer nombre.\n').upper()
if nombre in estudiantes['aprobados']:
    print(f'Tu {nombre} aprobaste la materia con {calificaciones_mate[nombre]}')
else:
    print(f'Tu {nombre} reprobaste la materia con {calificaciones_mate[nombre]}')
```
